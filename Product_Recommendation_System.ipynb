# **Model Based Collaborative Filtering System**

```python
#subset of rating
amazon_rating1=amazon_rating.head(10000)
amazon_rating1.shape
```

```python
#show pivot tables of ratings
ratingMatrix=amazon_rating1.pivot_table(index=['UserId'],columns=['ProductId'],values='Rating',fill_value=0)
ratingMatrix.head()
```

```python
#shape of rating
ratingMatrix.shape
```

```python
#Transposing the matrix
X=ratingMatrix.T
X.head()
```

```python
X.shape
```

```python
#unique product
X1=X
```

image.png

```python
#Decomposing the matrix
import sklearn
from sklearn.decomposition import TruncatedSVD

SVD=TruncatedSVD(n_components=10)
decomposed_matrix=SVD.fit_transform(X)
decomposed_matrix.shape
```

```python
decomposed_matrix
```

```python
#Correlation Matrix
import numpy as np
correlation_matrix=np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```

```python
prod_desc=pd.read_csv('product_descriptions.csv')
prod_desc.head()
```

```python
#Missing values
prod_desc.isnull().sum()
```

```python
#Feature extraction
vectorizer=TfidfVectorizer(stop_words='english')
X=vectorizer.fit_transform(prod_desc['product_description'])
X
```

```python
#fitting k-means
true_k=10
model=KMeans(n_clusters=true_k,init='k-means++',max_iter=100,n_init=1)
# model.fit(X)
```

```python
#predicting and visualise
y_pred=model.fit_predict(X)
plt.plot(y_pred,'.')
plt.show()
```

```python
#optimal cluster
tk=10
model=KMeans(n_clusters=true_k,init='k-means++',max_iter=100,n_init=1)
y_pred=model.fit(X)
print("Top terms per cluster:")
order_centroids=model.cluster_centers_.argsort()[:,::-1]
terms=vectorizer.get_feature_names_out()
for i in range(true_k):
  print("Cluster %d:" %i)
  for ind in order_centroids[i,:10]:
    print(' %s' %terms[ind])
```

```python
#predicting cluster based on key search
def disp_recomd(prod):
  Y=vectorizer.transform([prod])
  prediction=model.predict(Y)
  i=prediction[0]
  for ind in order_centroids[i,:10]:
    print(' %s' %terms[ind])
```

```python
#Keyword - cutting tool
disp_recomd('cutting tool')
```
