# **Linear Regression**

used to predict the value of a variable based on the value of another variable. The variable you want to predict is called the dependent variable. The variable you are using to predict the other variable's value is called the independent variable.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
#
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
#Load dataset
df = pd.read_csv('salarydataset.csv')
df.head()
```

```python
#InDependent Variable
x=df[['YearsExperience']]

#Dependent Variable
y=df['Salary']

#Fit the linear Regression
reg=LinearRegression().fit(x,y)

#Model score (R*2)
print("Score ",reg.score(x,y))

#Coefficient
print("Coefficient ",reg.coef_)

#Intercept
print("Intercept ",reg.intercept_)

#Predicting values
y_pred=reg.predict(x)
# print("Predicted values ",y_pred)

#plotting
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='red',label='Actual')
plt.plot(x,y_pred,color='blue',label='Fitted Line')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title("Years of Experince vs Salary")
plt.legend()
plt.show()
```

# **Polynomial Regression**

 form of regression analysis in which the relationship between the independent variable x and the dependent variable y is modeled as an nth degree polynomial in x.

```python
from sklearn.preprocessing import PolynomialFeatures

#InDependent Variable
x=df[['YearsExperience']]

#Dependent Variable
y=df['Salary']

#polynomialFeatures
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
# print(x_poly)

#Fit the linear Regression
reg=LinearRegression().fit(x_poly,y)
# print(reg.score(x_poly,y))

#Coefficient
print("Coefficient ",reg.coef_)

#Intercept
print("Intercept ",reg.intercept_)

#Predicting values
y_pred=reg.predict(x_poly)
# print("Predicted values ",y_pred


#plotting
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='red',label='Actual')
plt.plot(x,y_pred,color='blue',label='Polynomial')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title("Polynomial Regression :- Years of Experince vs Salary")
plt.legend()
plt.show()
```

# **Logistic Regression**

 a supervised machine learning algorithm widely used for binary classification tasks, such as identifying whether an email is spam or not and diagnosing diseases by assessing the presence or absence of specific conditions based on patient test results. use for Binary, probability


```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures

#InDependent Variable
x=df[['YearsExperience']]

#Dependent Variable
y=df['Salary']

#Let Salary into Binary outcome
#E.g if Salary > 50000 is 1 otherwise 0
y=(df['Salary']>50000).astype(int)

#Fit the Logistic Regressio model
reg=LogisticRegression().fit(x,y)
print("Score ",reg.score(x,y))

#Coefficient
print("Coefficient ",reg.coef_)

#Intercept
print("Intercept ",reg.intercept_)

#Predicting values
y_pred=reg.predict_proba(x)[:,1]
print("Predicted values ",y_pred)

#plotting
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='red',label='Actual')
plt.plot(x,y_pred,color='blue',label='Logistic Regression')
plt.xlabel('YearsExperience')
plt.ylabel('Probability of Salary > 50000')
plt.title("Logistic Regression :- Years of Experince vs Salary")
plt.legend()
plt.show()
```

# **Multiple Linear Regression**

 Multiple linear regression (MLR) is a statistical technique that uses several explanatory variables to predict the outcome of a response variable. It is also known as multiple regression, Multiple regression is an extension of linear (OLS) regression that uses just one explanatory variable.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression


```

```python
df=pd.read_excel('CarData.xlsx')
df.head()
```

```python
df.shape
print(df['Car'].unique())
```

```python
#Independent Variable
x=df[['Weight','Volume']]

#Dependent
y=df['CO2']

#Fit the linear Regression
reg=LinearRegression().fit(x,y)
print("Score ",reg.score(x,y))

#predict the co2 of a car where weight is 2300kg and volume is 1300
#Model score (R*2)

predictedCo2=reg.predict([[2300,1300]])
print("Predicted Co2 ",predictedCo2)

#predict the co2 of a car where weight is 1000kg and volume is 1000
predictedCo22=reg.predict([[1000,1000]])
print("Predicted Co2 ",predictedCo22)

#Coefficient
print("Coefficient ",reg.coef_)

#Intercept
print("Intercept ",reg.intercept_)

#visualise the regression fit for weight vs co2
sns.regplot(x='Weight',y='CO2',data=df)
plt.show()

#visualise the regression fit for volume vs co2
sns.regplot(x='Volume',y='CO2',data=df)
plt.show()

#pairplot to visulize relationships
sns.pairplot(df[['Weight','Volume','CO2']],kind='reg')
plt.show()
```
