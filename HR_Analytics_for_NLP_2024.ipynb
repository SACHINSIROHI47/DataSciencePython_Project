# **Stop Word Removal**
image.png

```python
nltk.download('stopwords')
stop_words=nltk.corpus.stopwords.words('english')
", ".join(stop_words)
```

**Remove the Stop Words**

```python
filtered_q0=[word for word in tok_q0 if word not in stop_words]
" ".join(filtered_q0)
```

Stemming
image.png

```python
#stemming
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

#write text for lemma
data="I am going to home"

words=word_tokenize(data)
ps=nltk.PorterStemmer()

for w in words:
  print(ps.stem(w),end=" ")
```

```python
msg=['change','changes','changing','changed','changer']
for w in msg:
  print(ps.stem(w),end=" ")
```

```python
#lemma
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
# print(lemmatizer.lemmatize("changes"))

for w in msg:
  print(lemmatizer.lemmatize(w),end=" ")

```

```python
#Lebelling
def rating(x):
  if x>3:
    return "Positive"
  elif x<3:
    return "Negative"
  else:
    return "Neutral"

df['Rating']=df['overall'].apply(rating)
df['Rating'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.show()
```
