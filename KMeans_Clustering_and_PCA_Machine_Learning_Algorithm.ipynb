# **K-Means**
 a machine learning algorithm that groups data points into clusters based on their distance from a cluster center


 image.png

image.png

```python
#k-means program
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x=np.array([[3,2],[1,4],[1,0],[10,4],[10,0],[6,5]])

#fit into k-mean
kmeans=KMeans(n_clusters=2,random_state=0,n_init=10).fit(x)
kmeans.labels_
kmeans.predict([[0,0],[12,3]])
kmeans.cluster_centers_
plt.scatter(x[:,0],x[:,1],c=kmeans.labels_,cmap='rainbow')
```

```python
#load_digit data for k-means
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

digits=load_digits()
print(digits.data)
print(digits.target)

#fit data into pca(Principle Component Analysis)

pca=PCA(n_components=2)
pca.fit(digits.data)

#transform the data
x=pca.transform(digits.data)
print(x)

plt.plot(x[:,0],x[:,1],'o')
plt.show()
```

```python
#use kmean cluster
kmeans=KMeans(n_clusters=10,random_state=0,n_init=10).fit(x)
kmeans.labels_
kmeans.predict([[0,0],[12,3]])
kmeans.cluster_centers_

plt.scatter(x[:,0],x[:,1],c=kmeans.labels_,cmap='rainbow')
plt.show()
```

```python
#finout centroid of kmeans
print(kmeans.cluster_centers_)

plt.scatter(x[:,0],x[:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.show()
```
