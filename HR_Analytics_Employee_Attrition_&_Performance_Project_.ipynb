```python
#plot age distribution data
fig,ax=plt.subplots(2,2,figsize=(10,10))

sns.distplot(df['Age'],ax=ax[0,0])
ax[0][0].set_title('Age Distribution')

sns.distplot(df['TotalWorkingYears'],ax=ax[0][1])
ax[0][1].set_title('Total Working Years')

sns.distplot(df['YearsAtCompany'],ax=ax[1][0])
ax[1][0].set_title('Years At Company')

sns.distplot(df['YearsInCurrentRole'],ax=ax[1][1])
ax[1][1].set_title('Years In Current Role')

plt.show()
```

```python
#Count Plot vertically
plt.figure(figsize=(10,3))
sns.countplot(df['Attrition'])
plt.show()
```

```python
#barplot gender wise attrition
plt.figure(figsize=(4,4))
#typecate attrition into int
sns.barplot(x='Gender',y='Attrition',data=df,hue='Gender')
plt.show()
```

```python
#Attrition accoring to montly income
plt.figure(figsize=(4,4))
sns.barplot(x='Attrition',y='MonthlyIncome',data=df,hue='Attrition')
plt.show()
```

```python
#job satisfaction according to monthly income
plt.figure(figsize=(4,4))
sns.barplot(x='JobSatisfaction',y='MonthlyIncome',data=df,hue='JobSatisfaction')
plt.show()
```

```python
#job satisfaction according to monthly income using boxplot chart
plt.figure(figsize=(4,4))
sns.boxplot(x='JobSatisfaction',y='MonthlyIncome',data=df,hue='JobSatisfaction')
plt.show()
```

```python
#job satisfaction according to monthly income using violin plot chart
plt.figure(figsize=(4,4))
sns.violinplot(x='JobSatisfaction',y='MonthlyIncome',data=df,hue='JobSatisfaction')
plt.show()
```

```python
#monthly income according to age using join plot (Scatter plot)
sns.jointplot(x='Age',y='MonthlyIncome',data=df,kind='scatter')
plt.show()
```

**Mutlivariante Analysis**

```python
#Multivariente analysis
data=['Attrition','Age','MonthlyIncome','DistanceFromHome']
sns.pairplot(df[data],kind='reg',diag_kind="kde",hue='Attrition')
plt.show()
```
