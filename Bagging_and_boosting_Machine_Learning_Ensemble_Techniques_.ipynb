# **Bagging and boosting Machine Learning Ensemble Techniques**

are both ensemble learning techniques that improve prediction accuracy by combining the predictions of multiple base learners:

**Bagging**

Trains multiple weak learners in parallel on different subsets of data. The predictions of these learners are then combined, usually by averaging or voting, to produce the final prediction. Bagging can be used with unstable models like decision trees.

**Boosting**

Trains weak learners sequentially, with each model learning from the errors of the previous one. Boosting assigns higher weights to instances misclassified by previous models, ensuring that subsequent models pay more attention to these problematic cases. Boosting works better for stable models like linear regression.

```python
# prompt: bagging and booting machine learning example

import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

X, y = make_classification(n_samples=100, n_features=4,
                           n_informative=2, n_redundant=0,
                           random_state=0, shuffle=False)
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a base Decision Tree classifier
base_classifier = DecisionTreeClassifier()

#apply the baggin boosting
bagging_classifier = BaggingClassifier(base_classifier, n_estimators=10, random_state=42)
boosting_classifier = AdaBoostClassifier(base_classifier, n_estimators=10, random_state=42)

#train the classifier
bagging_classifier.fit(X_train, y_train)
boosting_classifier.fit(X_train, y_train)

# Make predictions with the ensemble classifiers
y_pred_bagging = bagging_classifier.predict(X_test)
y_pred_boosting = boosting_classifier.predict(X_test)

#print accuracy
print(f"Bagging Classifier Accuracy: {accuracy_score(y_test, y_pred_bagging)}")
print(f"Boosting Classifier Accuracy: {accuracy_score(y_test, y_pred_boosting)}")

#plot the data after bagging and boosting
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred_bagging, cmap='coolwarm', alpha=0.8)
```
