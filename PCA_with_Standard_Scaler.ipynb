# **Principal component analysis (PCA)**
 is a statistical technique that reduces the number of dimensions in a dataset to make it easier to interpret:

image.png


# **What it does**

Transforms data into a smaller set of variables, called principal components, to simplify complexity and retain patterns
# **How it works**
Linearly transforms data onto a new coordinate system to identify directions with the largest variation

# **Applications**

Data preprocessing, machine learning algorithms, regression analysis, pattern recognition, signal processing, exploratory data analysis, building predictive models

```python
#load library for pca
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
#generate synthetic data
x,y=make_blobs(n_samples=100,n_features=2, random_state=42)
x.shape
```

```python
#train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#print with labels
print("x_train is ",x_train.shape)
print("y_train is ",y_train.shape)
print("x_test is ",x_test.shape)
print("y_test is ",y_test.shape)
```

```python
#convert to dataframe
df=pd.DataFrame(x_test,columns=["Feature1","Feature2"])
df.head()
```

# **PCA After Scaled**

```python
#standardize the data after train test split
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x_test)
```

```python
#apply the pca
pca=PCA()
pca.fit(x_scaled)
```

```python
#create the dataframe with pca
df_pca=pd.DataFrame(pca.transform(x_scaled),columns=["PC1","PC2"])
df_pca.head()
```

```python
#plot the pca result
plt.scatter(df_pca["PC1"],df_pca["PC2"])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Result")
plt.show()
```

# **PCA Before Scaled**

```python
#pca before scaled
pca_before=PCA()
pca_before.fit(x_test)
df_pca_before=pd.DataFrame(pca_before.transform(x_test),columns=["PC1","PC2"])
df_pca_before.head()
```

```python
#plot the pca results
plt.scatter(df_pca_before["PC1"],df_pca_before["PC2"])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Result Before Scaled")
plt.show()
```
